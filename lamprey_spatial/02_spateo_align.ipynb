```python
import warnings
warnings.filterwarnings('ignore')
import scanpy as sc
import pandas as pd
import spateo as st
```

```python
csv = pd.read_csv('/data/work/02.QC/final_x.txt')
# csv
```

```python
all_adata = []
for i in csv['path']:
    if 'h5ad' in i:
        slice_id = i.split('/')[-1].split('.')[0]
        adata = sc.read(i)
        adata.var = pd.DataFrame(index = adata.var.index)
        adata.obs['slices'] = slice_id
        adata.uns['__type'] = 'UMI'
        all_adata.append(adata)
        # adata_1.obs['batch'] = slice_id
        # l = [adata, adata_1]
        # adata =  sc.AnnData.concatenate(*l, join='outer')

        # print(i)
```

```python
aligned_slices, pis = st.align.morpho_align(
    models=[slice.copy() for slice in all_adata],
    spatial_key="spatial",
    key_added="align_spatial",
    device="0" # or device="cpu"
)
```
